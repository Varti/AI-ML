## Part A - Q2 
An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions. [4 Marks]

### 2.A. 
Probability that none of the items are defective? [1 Mark]
### 2.B. 
Probability that exactly one of the items is defective? [1 Mark]
### 2.C. 
Probability that two or fewer of the items are defective? [1 Mark]
### 2.D. 
Probability that three or more of the items are defective ? [1 Mark]

### Solution: Part A- Q2

We will use Binomial distribution \
probability of failure, p = 0.05 \
n= 10 

```python
n=10
p=0.05
k= np.arange(0,11)
```

```python
binomial= binom.pmf(n=n,p=p,k=k)
binomial
```

```python
#Q2 A : Probability that none of the items are defective?
binomial[0]
```

```python
#Q2 B :Probability that exactly one of the items is defective?
binomial[1]
```

```python
#Q2 C :Probability that two or fewer of the items are defective? Can also use cdf():binom.cdf(k=2, n=n, p=p)
binomial[0]+binomial[1]+binomial[2]
```

```python
#Q2 D :Probability that three or more of the items are defective?
1- (binomial[0]+binomial[1]+binomial[2])
```

##### Q4 B: Write the null hypothesis and alternative hypothesis.

In testing if the funds raised by companies are significantly different by Closed & operating companies, the null hypothesis states that the mean of funds raised by both the Operating & Closed companies are same,  𝜇𝐴 = 𝜇𝐵. 

The alternate hypothesis states that mean of funds raised by both the Operating & Closed companies are different, 𝜇𝐴 ≠ 𝜇𝐵.

Null Hypothesis (𝐻0) : 𝜇𝐴 = 𝜇𝐵
Alternate Hypothesis (HA) : 𝜇𝐴 ≠ 𝜇𝐵


##### Q4 C: Test for significance and conclusion

Here, we are selecting significance level (𝛼) = 0.05

```python
#Calculate test statistics & p-value
t_statistic, p_value  = ttest_ind(sample_closed['Funds_in_million'],sample_operating['Funds_in_million'])
print('tstat',t_statistic)    
print('P Value',p_value)   
```

#### Conclusion
If p-value is less than significance level, then we can reject the null hypothesis.
